```python
players = pd.read_csv('C:/Users/User/PycharmProjects/scraping/NBA_MVP_PROJECT/player/players.csv')
```

```python
del players['Unnamed: 0']
del players['Rk']
```

```python
players.head()
```

```python
players['Player'].head(50)
# There are some asterisk after the players name
# We have to clean to be able to merge the dataframes according to the name and year
```

```python
# * have a meaning for regular expressions, so we have to put regex to consider '*' a normal string 
players['Player'] = players['Player'].str.replace('*','', regex=False)
```

```python
players.head(50)
```

```python
players.head(20)
```

```python
# Some players have more than one row cause they played in more then one team
# We need just one to run the ML model
players.groupby(['Player', 'year'])
```

```python
def single_row(df):
    if df.shape[0]==1:
        return df
    else:
        row = df[df['Tm'] == 'TOT']
        row['Tm'] = df.iloc[-1,:]['Tm']
        return row
players = players.groupby(['Player', 'year']).apply(single_row)
```

```python
players.head(20)
```

```python
# Run this command 2 times to take the excess of index in the dataframe
players.index = players.index.droplevel()
```

```python
players
```

```python
# We use outer merge cause not all of the rows are in both dfs
# 'If you can't find the player, just keep the info'
# Inner would be deleting all the rows not found
combined = players.merge(mvps, how='outer', on=['Player', 'year'])
```

```python
combined.head(5)
```

```python
combined[combined['Pts Won'] > 0]
```

```python
combined[['Pts Won', 'Pts Max', 'Share']] = combined[['Pts Won', 'Pts Max', 'Share']].fillna(0)
```

```python
teams = pd.read_csv('C:/Users/User/PycharmProjects/scraping/NBA_MVP_PROJECT/team/teams.csv')
```

```python
# Cleaning the additional rows with Division information
# Remove the row that contains the word division on the 'w' column
teams = teams[~teams['W'].str.contains('Division')].copy()
```

```python
# Removing * from team names
teams['Team'] = teams['Team'].str.replace('*','', regex=False)
```

```python
# lines[1:] escape the first row(header)
nicknames = {}
with open('C:/Users/User/PycharmProjects/scraping/NBA_MVP_PROJECT/team/nicknames.csv', encoding='utf-8') as f:
    lines = f.readlines()
    for line in lines[1:]:
        abbrev, name = line.replace('\n','').split(',')
        nicknames[abbrev] = name
```

```python
combined['Team'] = combined['Tm'].map(nicknames)
```
